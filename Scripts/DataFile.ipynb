# Create File with Storm Indexes, Temperatures, and Pressures of All Stations

```python
import storm_indices as si
import pandas as pd
```

```python
TP = pd.read_csv('../Data/scottbase/scottbase.csv')
TP['DateTime'] = pd.to_datetime(TP['DateTime'],format='%Y-%m-%d')
TP.index = TP['DateTime']
del TP['DateTime']
TP = TP[['Mean Temperature [Deg C]', 'Mean sea level pressure [Hpa]']]
TP = TP[~TP.index.isna()]
TP = si.compute_si(TP, 'Mean Temperature [Deg C]', 'Mean sea level pressure [Hpa]', 'SI')
TP = TP.rename(columns={'Mean Temperature [Deg C]': 'SB_T', 'Mean sea level pressure [Hpa]': 'SB_P', 'SI': 'SB_SI'})
SB = TP.copy()

TP = pd.read_csv('../Data/marblepoint/marblepoint.csv')
TP['DateTime'] = pd.to_datetime(TP['DateTime'],format='%Y-%m-%d')
TP.index = TP['DateTime']
del TP['DateTime']
TP = TP[['temperature', 'pressure']]
TP = TP[~TP.index.isna()]
TP = si.compute_si(TP, 'temperature', 'pressure', 'SI')
TP = TP.rename(columns={'temperature': 'MP_T', 'pressure': 'MP_P', 'SI': 'MP_SI'})
MP = TP.copy()

TP = pd.read_csv('../Data/capebird/capebird.csv')
TP['DateTime'] = pd.to_datetime(TP['DateTime'],format='%Y-%m-%d')
TP.index = TP['DateTime']
del TP['DateTime']
TP = TP[['temperature', 'pressure']]
TP = TP[~TP.index.isna()]
TP = si.compute_si(TP, 'temperature', 'pressure', 'SI')
TP = TP.rename(columns={'temperature': 'CB_T', 'pressure': 'CB_P', 'SI': 'CB_SI'})
CB = TP.copy()

TP = pd.read_csv('../Data/margaret/margaret.csv')
TP['DateTime'] = pd.to_datetime(TP['DateTime'],format='%Y-%m-%d')
TP.index = TP['DateTime']
del TP['DateTime']
TP = TP[['temperature', 'pressure']]
TP = TP[~TP.index.isna()]
TP = si.compute_si(TP, 'temperature', 'pressure', 'SI')
TP = TP.rename(columns={'temperature': 'MR_T', 'pressure': 'MR_P', 'SI': 'MR_SI'})
MR = TP.copy()

TP = pd.read_csv('../Data/vito/vito.csv')
TP['DateTime'] = pd.to_datetime(TP['DateTime'],format='%Y-%m-%d')
TP.index = TP['DateTime']
del TP['DateTime']
TP = TP[['temperature', 'pressure']]
TP = TP[~TP.index.isna()]
TP = si.compute_si(TP, 'temperature', 'pressure', 'SI')
TP = TP.rename(columns={'temperature': 'VT_T', 'pressure': 'VT_P', 'SI': 'VT_SI'})
VT = TP.copy()

TP = pd.read_csv('../Data/ferrell/ferrell.csv')
TP['DateTime'] = pd.to_datetime(TP['DateTime'],format='%Y-%m-%d')
TP.index = TP['DateTime']
del TP['DateTime']
TP = TP[['temperature', 'pressure']]
TP = TP[~TP.index.isna()]
TP = si.compute_si(TP, 'temperature', 'pressure', 'SI')
TP = TP.rename(columns={'temperature': 'FR_T', 'pressure': 'FR_P', 'SI': 'FR_SI'})
FR = TP.copy()

TP = pd.read_csv('../Data/manuela/manuela.csv')
TP['DateTime'] = pd.to_datetime(TP['DateTime'],format='%Y-%m-%d')
TP.index = TP['DateTime']
del TP['DateTime']
TP = TP[['temperature', 'pressure']]
TP = TP[~TP.index.isna()]
TP = si.compute_si(TP, 'temperature', 'pressure', 'SI')
TP = TP.rename(columns={'temperature': 'MN_T', 'pressure': 'MN_P', 'SI': 'MN_SI'})
MN = TP.copy()

TP = pd.read_csv('../Data/alessandra/alessandra.csv')
TP['DateTime'] = pd.to_datetime(TP['DateTime'],format='%Y-%m-%d')
TP.index = TP['DateTime']
del TP['DateTime']
TP = TP[['temperature_C', 'pressure_hPa']]
TP = TP[~TP.index.isna()]
TP = si.compute_si(TP, 'temperature_C', 'pressure_hPa', 'SI')
TP = TP.rename(columns={'temperature_C': 'AL_T', 'pressure_hPa': 'AL_P', 'SI': 'AL_SI'})
AL = TP.copy()

TP = pd.read_csv('../Data/silvia/silvia.csv')
TP['DateTime'] = pd.to_datetime(TP['DateTime'],format='%Y-%m-%d')
TP.index = TP['DateTime']
del TP['DateTime']
TP = TP[['temperature_C', 'pressure_hPa']]
TP = TP[~TP.index.isna()]
TP = si.compute_si(TP, 'temperature_C', 'pressure_hPa', 'SI')
TP = TP.rename(columns={'temperature_C': 'SV_T', 'pressure_hPa': 'SV_P', 'SI': 'SV_SI'})
SV = TP.copy()
```
