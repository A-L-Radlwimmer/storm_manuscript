```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.image as mpimg
import os
import numpy as np
```

```python
fig = plt.figure(figsize=(18, 16))
gs = gridspec.GridSpec(1, 2, figure=fig, width_ratios=[8, 5], wspace=0., hspace=0.)
fig.subplots_adjust(hspace=-0.5)

ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1])
axes = [ax1, ax2]

image_dir = "../Figures"
list_of_subplots = ['all_si_and_ice.png', 'Bild1.png']
path_list = [os.path.join(image_dir, i) for i in list_of_subplots]

for ax, img_path in zip(axes, path_list):
    img = mpimg.imread(img_path)
    ax.imshow(img)
    ax.axis('off')
ax2.set_title(r'$\bf{(b)}$ Formation of fast ice in McMurdo Sound', fontsize=16, x=0.01, y=1.03, ha='left')

plt.savefig('../Figures/all_si_and_fast_ice_formation', bbox_inches='tight')
plt.show()
```

```python
fig = plt.figure(figsize=(16, 18))
gs = gridspec.GridSpec(2, 1, figure=fig, height_ratios=[1, 1], wspace=0., hspace=0.1)
fig.subplots_adjust(hspace=-0.5)

ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1])
axes = [ax1, ax2]

image_dir = "../Figures"
list_of_subplots = ['ScottBase_3year_std.png', 'RossSea_polynyas_and_ScottBase_SI_3_year_smoothing_window_new.png']
path_list = [os.path.join(image_dir, i) for i in list_of_subplots]

for ax, img_path in zip(axes, path_list):
    img = mpimg.imread(img_path)
    ax.imshow(img)
    ax.axis('off')
ax1.set_title(r'$\bf{(a)}$ Inter-annual variability of storm index', fontsize=24, x=0.01, y=1.03, ha='left')
ax2.set_title(r'$\bf{(b)}$ Low-frequency variation of storm index, polynya activity, and SAM', fontsize=24, x=0.01, y=1.03, ha='left')

plt.savefig('../Figures/std_and_sam', bbox_inches='tight')
plt.show()
```

```python
fig = plt.figure(figsize=(15, 15))
gs = gridspec.GridSpec(4, 3, figure=fig, height_ratios=[1, 1, 1, 1], width_ratios=[1, 1, 1], hspace=0., wspace=0.)

ax1 = fig.add_subplot(gs[2:4, 0:2])  # Scott Base
ax2 = fig.add_subplot(gs[3, 2])      # Margaret
ax3 = fig.add_subplot(gs[2, 2])      # Vito
ax4 = fig.add_subplot(gs[1, 2])      # Ferrell
ax5 = fig.add_subplot(gs[1, 1])      # Cape Bird
ax6 = fig.add_subplot(gs[1, 0])      # Marble Point
ax7 = fig.add_subplot(gs[0, 0])      # Manuela
ax8 = fig.add_subplot(gs[0, 1])      # Alessandra
ax9 = fig.add_subplot(gs[0, 2])      # Silvia
axes = [ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9]

image_dir = "../Figures"
list_of_subplots = ['ScottBase_asi_summer.png', 'Margaret_asi_summer.png', 'Vito_asi_summer.png', 'Ferrell_asi_summer.png',
                    'CapeBird_asi_summer.png', 'MarblePoint_asi_summer.png',
                    'Manuela_asi_summer.png', 'Alessandra_asi_summer.png', 'Silvia_asi_summer.png']
path_list = [os.path.join(image_dir, i) for i in list_of_subplots]

for ax, img_path in zip(axes, path_list):
    img = mpimg.imread(img_path)
    ax.imshow(img)
    ax.axis('off')

plt.savefig('../Figures/all_asi_summer', bbox_inches='tight')
plt.tight_layout()
plt.show()
```
