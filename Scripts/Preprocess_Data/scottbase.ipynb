### Combine Temperature and Pressure Data from Scott Base into CSV-File

```python
import pandas as pd
import pytz
```

```python
temperature_df_6237 = pd.read_csv('../../Data/scottbase/6237__Temperature__daily.csv')
#temperature_df_6237 = temperature_df_6237.iloc[:17000] # Last years seem unreliable
temperature_df_12740 = pd.read_csv('../../Data/scottbase/12740__Temperature__daily.csv')
temperature_df_44557 = pd.read_csv('../../Data/scottbase/44557__Temperature__daily.csv')
temperature_df = pd.concat([temperature_df_6237, temperature_df_12740, temperature_df_44557], ignore_index=True)
temperature_df['DateTime'] = pd.to_datetime(temperature_df['Observation time UTC'])
temperature_df.index = temperature_df['DateTime']
temperature_df.index = temperature_df.index.tz_convert(None)
temperature_df.index = temperature_df.index.floor('D')
temperature_df = temperature_df[~temperature_df.index.duplicated(keep='first')]

temperature_df
```

```python
pressure_df_6237 = pd.read_csv('../../Data/scottbase/6237__Pressure__daily.csv')
#pressure_df_6237 = pressure_df_6237.iloc[:17000] # Last years seem unreliable

pressure_df_44557 = pd.read_csv('../../Data/scottbase/44557__Pressure__hourly.csv')
pressure_df_44557['DateTime'] = pd.to_datetime(pressure_df_44557['Observation time UTC'], utc=True)
nz_tz = pytz.timezone('Pacific/Auckland')
pressure_df_44557['DateTime_NZ'] = pressure_df_44557['DateTime'].dt.tz_convert(nz_tz)
pressure_df_44557 = pressure_df_44557[pressure_df_44557['DateTime_NZ'].dt.hour == 9]
pressure_df_44557 = pressure_df_44557.reset_index()
del pressure_df_44557['DateTime']
del pressure_df_44557['DateTime_NZ']

pressure_df_12740 = pd.read_csv('../../Data/scottbase/12740__Pressure__hourly.csv')
pressure_df_12740['DateTime'] = pd.to_datetime(pressure_df_12740['Observation time UTC'], utc=True)
nz_tz = pytz.timezone('Pacific/Auckland')
pressure_df_12740['DateTime_NZ'] = pressure_df_12740['DateTime'].dt.tz_convert(nz_tz)
pressure_df_12740 = pressure_df_12740[pressure_df_12740['DateTime_NZ'].dt.hour == 9]
pressure_df_12740 = pressure_df_12740.reset_index()
del pressure_df_12740['DateTime']
del pressure_df_12740['DateTime_NZ']

pressure_df = pd.concat([pressure_df_12740, pressure_df_6237, pressure_df_44557], ignore_index=True)
pressure_df['DateTime'] = pd.to_datetime(pressure_df['Observation time UTC'])
pressure_df.index = pressure_df['DateTime']
pressure_df.index = pressure_df.index.tz_convert(None)
pressure_df.index = pressure_df.index.floor('D')
pressure_df = pressure_df[~pressure_df.index.duplicated(keep='first')]

pressure_df
```

```python
TP = pd.concat([temperature_df['Mean Temperature [Deg C]'], pressure_df['Mean sea level pressure [Hpa]']], axis=1)
TP
```

```python
TP.to_csv('../../Data/scottbase/scottbase.csv')
```
